<h3/>Question 5</h3>

<h3>Insight 5:</h3>

<h3>Question 6</h3>
What is the range of bill length for Gentoo penguins?

```python
penguins[penguins['species'] == 'Gentoo']['bill_length_mm'].max() - penguins[penguins['species'] == 'Gentoo']['bill_length_mm'].min()
```

<h3>Insight 6</h3>

<h3>Question 7</h3>
